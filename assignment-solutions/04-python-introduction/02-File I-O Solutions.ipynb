# File I/O Operations

    **1. Write a Python program to count the frequency of words in the text file "Python-wiki.txt".

```python
#!/usr/bin/python
import collections

wordcount = collections.Counter()
with open("..\data/Python-wiki.txt") as file:
    for line in file:
        wordcount.update(line.split())

for k,v in wordcount.items():
    print(k, v)
```

```python
with open("..\data\Python-wiki.txt", "r") as file: 
    data = file.readlines() 
    for line in data: 
        word = line.split() 
        print(word) 
```

** 2. Write a Python program to read first n lines of the file "zen-of-python.txt".

```python
def file_read_from_head(fname, nlines):
        from itertools import islice
        with open(fname) as f:
                for line in islice(f, nlines):
                        print(line)
file_read_from_head('..\data\zen-of-python.txt',2)
```

```python
def file_read_from_head(fname, nlines):
    count = 0
    with open(fname) as f:
        for line in f.readlines():
            if count < nlines:
                print(line)
                count+=1
            else:
                return
            
file_read_from_head('..\data\zen-of-python.txt',2)
```

** 3. Write a Python program to remove newline characters from a file.

```python
def remove_newlines(fname):
    flist = open(fname).readlines()
    return [s.rstrip('\n') for s in flist]

print(remove_newlines("..\data\zen-of-python.txt"))
```
