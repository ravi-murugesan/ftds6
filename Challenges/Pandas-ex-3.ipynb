Attribute Information:

- thalach: maximum heart rate achieved
- exang: exercise induced angina

- slope: the slope of the peak exercise ST segment
- ca: number of major vessels (0-3) colored by flourosopy

4. For the column cp, map the old values to new values as follows:

    - 0 --> no chest pain
    - 1 --> moderate chest pain
    - 2 --> serious chest pain
    - 3 --> severe chest pain

```python
df['cp'] = df['cp'].replace({0 : 'no chest pain', 1: 'moderate chest pain', 
                             2: 'serious chest pain', 3: 'severe chest pain'})
df.head(5)
```

5. Convert the categorical column cp to dummy variables.

```python
df = pd.get_dummies(data=df, columns=['cp'])
df.head(5)
```

6. Create a new column called bp_risk that maps the values in the column trestbps to the values as follows:

    - less than 90 --> low
    - 90-119 --> ideal
    - 120-139 --> at-risk
    - more than 140 --> high

```python
df['bp_risk'] = ['low' if i < 90 else 'ideal' if 90 <= i <= 119 else
                 'at-risk' if 120<= i <= 139 else 'high' for i in df['trestbps']]
df.head(5)
```

7. Create a pie chart that plots out the column bp_risk.

```python
df1 = df['bp_risk'].value_counts()
plot = df1.plot.pie(figsize=(5, 5),label =None)
```
